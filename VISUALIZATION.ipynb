***Exploratory Data Analysis***

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_with_markdown = pd.read_csv('data_with_markdown.csv')
dataset_no_markdown = pd.read_csv('data_no_mark_down.csv')
```

```python
dataset_with_markdown = dataset_with_markdown.drop(['Unnamed: 0'], axis=1)
dataset_no_markdown = dataset_no_markdown.drop(['Unnamed: 0'], axis=1)
```

```python
a = dataset_with_markdown.query('Year==2010')['Month'].value_counts()
b = dataset_with_markdown.query('Year==2011')['Month'].value_counts()
c = dataset_with_markdown.query('Year==2012')['Month'].value_counts()
d = dataset_with_markdown.query('Year==2013')['Month'].value_counts()

print('2010------------------------------')
print(a)
print('2011------------------------------')
print(b)
print('2012------------------------------')
print(c)
print('2013------------------------------')
print(d)
```

```python
sns.set_style("white")
sns.set_context("notebook")

walmart_color = ['#0072D8', '#FFC220']
walmart_light = ['#85C6FF', '#FFE499']
walmart_blue = ['#0072D8', '#85C6FF']
walmart_orange = ['#FFC220', '#FFE499']
```

```python
# histplot
dataset_no_markdown.hist(bins=50, figsize=(20,15), color='#85C6FF')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('Effect of holiday on Weekly Sales in different months')

months = ['J', 'F', 'M', 'A', 'My', 'Jn', 'Jl', 'Au', 'S', 'O', 'N', 'D']

sns.barplot(ax=axes[0], data=dataset_no_markdown.query('Year==2010'), x="Month", y="Weekly_Sales", hue="IsHoliday", order = months, palette=walmart_light)
axes[0].set_title('2010')
sns.barplot(ax=axes[1], data=dataset_no_markdown.query('Year==2011'), x="Month", y="Weekly_Sales", hue="IsHoliday", order = months, palette=walmart_light)
axes[1].set_title('2011')
sns.barplot(ax=axes[2], data=dataset_no_markdown.query('Year==2012'), x="Month", y="Weekly_Sales", hue="IsHoliday", order = months, palette=walmart_light)
axes[2].set_title('2012')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10), sharey=True)
fig.suptitle('Effect of different variables on Weekly Sales', fontsize=15)
fig.subplots_adjust(hspace = 0.4)

sns.regplot(ax=axes[0, 0], x='CPI', y="Weekly_Sales", data=dataset_no_markdown, ci=None, color='#85C6FF', scatter_kws={'alpha':0.05}, line_kws={'color': '#FFC220'})
axes[0, 0].set_title('CPI', fontsize=17)
axes[0, 0].set(xlabel=None)

sns.regplot(ax=axes[0, 1], x="Unemployment", y="Weekly_Sales", data=dataset_no_markdown, ci=None, color='#85C6FF', scatter_kws={'alpha':0.05}, line_kws={'color': '#FFC220'})
axes[0, 1].set_title('Unemployment', fontsize=17)
axes[0, 1].set(xlabel=None)

sns.regplot(ax=axes[1, 0], x="Temperature", y="Weekly_Sales", data=dataset_no_markdown, ci=None, color='#85C6FF', scatter_kws={'alpha':0.05}, line_kws={'color': '#FFC220'})
axes[1, 0].set_title('Temperature', fontsize=17)
axes[1, 0].set(xlabel=None)

sns.regplot(ax=axes[1, 1], x="Fuel_Price", y="Weekly_Sales", data=dataset_no_markdown, ci=None, color='#85C6FF', scatter_kws={'alpha':0.05}, line_kws={'color': '#FFC220'})
axes[1, 1].set_title('Fuel Price', fontsize=17)
axes[1, 1].set(xlabel=None)
```
